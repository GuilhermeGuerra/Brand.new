````python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Rogerio', 'Arthur', 'Gabriela', 'Guilherme'],
    'salario': [np.nan, 5000, np.nan, 6000, 11000]
})

#calcula a média e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
'''#calcula a mediana e substitui os nulos pela mediana'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
````

````python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
````

````python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['Preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
````

````python
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Rogerio', 'Arthur', 'Gabriela', 'Guilherme'],
    'cidade': ['Rio de Janeiro', np.nan, 'São Paulo', 'Campinas', 'Rio de Janeiro']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
display(df_cidades)
````

````python
import pandas as pd
import plotly.express as px
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df_desafio = pd.read_excel("https://github.com/GuilhermeGuerra/Brand.new/raw/main/Salarios_Data_Scientist_Por_Pais.xlsx")
display(df_desafio.head(20))
````

````python
df_media = df_desafio[['Salário Mínimo (USD)', 'Salário Máximo (USD)']].mean(axis=1).to_frame(name='Média Salarial')
df_media['Média Salarial'] = df_media['Média Salarial'].astype('int64')
df_media.head()
````

````python
df_desafio = pd.concat([df_desafio, df_media], axis=1)
df_desafio.head()
````

````python
df_desafio.columns
````

````python
df_desafio.columns = ['País', 'Salário Mínimo (USD)', 'Salário Máximo (USD)',
       'Média Salarial (USD)']
````

````python
df_desafio.head()
````

````python
plt.figure(figsize=(40, 8))
sns.barplot(data=df_desafio, x='País', y='Média Salarial (USD)')
plt.title('Salário médio (Data Scientist)',fontsize=30)
plt.xlabel('Países',fontsize=20)
plt.ylabel('Salário médio (USD)',fontsize=20)
plt.show()
````

````python
df_desafio['País'] = df_desafio['País'].replace('Reino Unido', 'UK')
plt.figure(figsize=(40, 8))
sns.barplot(data=df_desafio, x='País', y='Média Salarial (USD)')
plt.title('Salário médio (Data Scientist)',fontsize=30)
plt.xlabel('Países',fontsize=25)
plt.ylabel('Salário médio (USD)',fontsize=25)
plt.show()
````

````python
df_desafio.groupby('País')['Média Salarial (USD)'].mean().sort_values(ascending=False)
````

````python
ordem = df_desafio.groupby('País')['Média Salarial (USD)'].mean().sort_values(ascending=True).index
````

````python
ordem
````

````python
plt.figure(figsize=(48, 15))
sns.barplot(data=df_desafio, x='País', y='Média Salarial (USD)',order=ordem, palette='viridis', hue='Média Salarial (USD)', legend=False)
plt.title('Salário médio (Data Scientist)',fontsize=30)
plt.xlabel('Países',fontsize=25)
plt.ylabel('Salário médio (USD)',fontsize=25)
plt.show()
````
